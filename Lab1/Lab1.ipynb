# L1: Lexical Analysis
## 1. Text Segmentation
Exo1: 

	INPUT=Nepals Home Minister Ramesh Lekhak resigned Monday following the violence, Communications Minister Prithvi Subba confirmed to reporters late Monday.
	OUTPUT=['Nepal', '', 's', 'Home', 'Minister', 'Ramesh', 'Lekhak', 'resigned', 'Monday', 'following', 'the', 'violence', ',', 'Communications', 'Minister', 'Prithvi', 'Subba', 'confirmed', 'to', 'reporters', 'late', 'Monday', '.']

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Nepals Home Minister Ramesh Lekhak resigned Monday following the violence, Communications Minister Prithvi Subba confirmed to reporters late Monday."

doc = nlp(text)
tokens = [token.text for token in doc]
tokens
```

## 2. Lemmatization

```python
lemmatized = [(token.text, token.lemma_) for token in doc]
lemmatized
```

## 3. Part-of-speech tagging 

```python
pos_tags = [(token.text, token.pos_) for token in doc]
pos_tags
```

# L2: Syntactic Analysis

```python
dependencies = [(token.text, token.dep_, token.head.text) for token in doc]
dependencies
```

# L3: Semantic Analysis
Exo2: 

	INPUT=Nepals Home Minister Ramesh Lekhak resigned Monday following the violence, Communications Minister Prithvi Subba confirmed to reporters late Monday.
	OUTPUT=[Nepal, Home Minister Ramesh Lekkak, Monday, Communications Minister Prithvi Subba, Monday]

```python
entities = [(ent.text, ent.label_) for ent in doc.ents]
entities
```

```python
important_entities = [ent.text for ent in doc.ents 
                      if ent.label_ in ["PERSON", "ORG", "GPE", "DATE"]]
important_entities
```

Exo3: 

	INPUT=The police are firing indiscriminately, one protester told Indian news agency ANI. At least 17 people were killed in Kathmandu and two more in the eastern city of Itahari, according to hospital officials. 
	OUTPUT=[The, Indian, ANI, At, 17, Kathmandu, Itahari]

```python
text2 = """The police are firing indiscriminately, one protester told Indian news agency ANI.
At least 17 people were killed in Kathmandu and two more in the eastern city of Itahari."""

doc2 = nlp(text2)

keywords = set()
for ent in doc2.ents:
    if ent.label_ in ["GPE", "ORG", "CARDINAL", "NORP"]:
        keywords.add(ent.text)

keywords
```

Exo3: 

	INPUT=D kin, s thu ngn sch t thu thu nhp c nhn cn khong 84.477 t ng, gim 21.000 t mi nm so vi hin hnh. S ngi cn phi np thu l 2,21 triu ngi, gim 2,18 triu ngi (49,66%) do chuyn t bc 1 sang din khng chu thu.
	OUTPUT=[D, 84.477, 21.000,  S, "2,21", "2,18",  49,66%]

```python
from underthesea import ner, word_tokenize

text_vi = """
D kin, s thu ngn sch t thu thu nhp c nhn cn khong 84.477 t ng,
gim 21.000 t mi nm so vi hin hnh. S ngi cn phi np thu l 2,21 triu ngi,
gim 2,18 triu ngi (49,66%)
"""

tokens_vi = word_tokenize(text_vi)
ner_vi = ner(text_vi)

numbers = [w for w in tokens_vi if any(c.isdigit() for c in w)]
keywords_vi = ["D"] + numbers + ["S"]

keywords_vi
```

# L4: Pragmatic Analysis

```python
def pragmatic_analysis(text):
    if "resigned" in text.lower():
        return "Political resignation event"
    if "killed" in text.lower():
        return "Violence / Casualties report"
    if "thu" in text.lower():
        return "Economic policy announcement"
    return "General information"

texts = [text, text2, text_vi]

for t in texts:
    print(pragmatic_analysis(t))
```
